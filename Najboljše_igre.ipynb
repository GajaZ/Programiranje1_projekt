Iz spletne strani SteamSpy (http://steamspy.com/) in GameRankings (http://www.gamerankings.com/sites/1188-ign/index.html?platform=19) poberemo podatke o PC igrah in jih spravimo v csv datoteke.

```python
import pandas as pd
import numpy as np
%matplotlib
```

```python
igre_zanri = pd.read_csv('Steam\zdruzeno_zanri.csv')
igre_podjetja = pd.read_csv('Steam\steam_distribucije\zdruzeno_podjetja.csv')
igre_ign = pd.read_csv('IGN\ign_ocene.csv')
```

# Analiza

```python
Najprej bom primerjala ocene ljudi, ki so prijavljeni na Steam-u, z ocenami tistih, ki se s poznavanjem PC iger profesionalno ukvarjajo.
```

```python
ocene_steam = igre_zanri[['ime', 'ocena_steam']]
skupne_ocene = pd.merge(ocene_steam, igre_ign, on=['ime'])
skupne_ocene.loc[skupne_ocene['ocena_ign'] > 0, 'ocena_ign'] = (skupne_ocene['ocena_ign'] * 10)
skupne_ocene.drop_duplicates().reset_index()
```

Iz spodnjega grafa se vidi, da so ocene kritikov (IGN) v povprečju veliko boljše od ocen igralcev (Steam)

```python
skupne_ocene1 = skupne_ocene[np.isfinite(skupne_ocene['ocena_steam']) & np.isfinite(skupne_ocene['ocena_ign'])]
skupne_ocene1.plot()
```

V spodnji tabeli je izračunana razlika med ocenami na Steam-u in tistimi na IGN-u.

```python
skupne_ocene2 = skupne_ocene1.sort_values(['ocena_steam', 'ocena_ign'], ascending = [True, False])
skupne_ocene2['absolutno'] = abs(skupne_ocene2['ocena_steam'] - skupne_ocene2['ocena_ign'])
skupne_ocene2.sort_values(['absolutno'],ascending= False)
```
